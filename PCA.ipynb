```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from google.colab import drive
drive.mount('/content/drive')

file = "/content/drive/MyDrive/Civilian Cyber Warrior Project/merge_wc_edit_update.csv"
df = pd.read_csv(file)
df.info()
```

```python
#PCA for Emotive National Identity
PCA1 = df[["Q36_2", "Q36_3", "Q36_4", "Q36_5", "Q36_7", "Q36_8"]]
PCA1 = PCA1.fillna(0)
PCA1 = np.array(PCA1)
x=PCA1
x=StandardScaler().fit_transform(x)
pca = PCA(n_components = 1)
principalComponent1 = pca.fit_transform(x)
df['Principal_Component1'] = principalComponent1
print(pca.explained_variance_ratio_)
```

```python
#PCA for Belief in Group Equality
PCA2 = df[["Q37_4", "Q37_5", "Q37_6", "Q37_7", "Q37_8", "Q37_9", "Q37_10"]]
PCA2 = PCA2.fillna(0)
PCA2 = np.array(PCA2)
x=PCA2
x=StandardScaler().fit_transform(x)
pca = PCA(n_components = 1)
principalComponent2 = pca.fit_transform(x)
df['Principal_Component2'] = principalComponent2
print(pca.explained_variance_ratio_)
```

```python
#PCA for Outgroup Antagonism
PCA3 = df[["Q36_17", "Q36_21", "Q37_1", "Q37_2", "Q37_3"]]
PCA3 = PCA3.fillna(0)
PCA3 = np.array(PCA3)
x=PCA3
x=StandardScaler().fit_transform(x)
pca = PCA(n_components = 1)
principalComponent3 = pca.fit_transform(x)
df['Principal_Component3'] = principalComponent3
print(pca.explained_variance_ratio_)
```
